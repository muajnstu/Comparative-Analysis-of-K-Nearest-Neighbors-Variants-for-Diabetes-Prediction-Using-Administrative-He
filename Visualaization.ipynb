```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Data with G-Mean added
data = {
    'Model Name': [
        'KNN', 'DistanceKNN', 'GeneralizedKNN', 'EuclideanKNN', 'ManhattanKNN',
        'ChebyshevKNN', 'MahalanobisKNN', 'SeuclideanKNN', 'WminkowskiKNN'
    ],
    'Accuracy': [0.9875, 0.9890, 0.9891, 0.9875, 0.9836, 0.9824, 0.9950, 0.9952, 0.9891],
    'Sensitivity': [0.9875, 0.9890, 0.9891, 0.9875, 0.9836, 0.9824, 0.9950, 0.9952, 0.9891],
    'F1 Score': [0.9870, 0.9885, 0.9887, 0.9870, 0.9823, 0.9818, 0.9949, 0.9951, 0.9887],
    'G-Mean': [0.9935, 0.9943, 0.9943, 0.9935, 0.9915, 0.9908, 0.9974, 0.9975, 0.9943],
    'AUC': [0.9964, 0.9964, 0.9967, 0.9964, 0.9949, 0.9953, 0.9985, 0.9987, 0.9967]
}

df = pd.DataFrame(data)

def compute_rpi(series):
    return (series - series.min()) / (series.max() - series.min()) * 100

metrics = ['Accuracy', 'Sensitivity', 'F1 Score', 'G-Mean', 'AUC']
for metric in metrics:
    df[f'RPI_{metric}'] = compute_rpi(df[metric])

bar_width = 0.18
x = np.arange(len(df['Model Name']))

plt.figure(figsize=(13,6))
plt.bar(x - 1.5*bar_width, df['RPI_Accuracy'], width=bar_width, color='grey', label='Accuracy')
plt.bar(x - 0.5*bar_width, df['RPI_Sensitivity'], width=bar_width, color='white', edgecolor='black', hatch='//', label='Sensitivity')
plt.bar(x + 0.5*bar_width, df['RPI_F1 Score'], width=bar_width, color='black', label='F1 Score')
plt.bar(x + 1.5*bar_width, df['RPI_G-Mean'], width=bar_width, color='dodgerblue', label='G-Mean')
plt.bar(x + 2.5*bar_width, df['AUC'], width=bar_width, color='orange', label='AUC')

plt.xlabel('Model Name')
plt.ylabel('RPI Score (%)')
plt.xticks(x, df['Model Name'], rotation=45)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.14), ncol=5, frameon=False)
plt.tight_layout()
plt.show()
```

```python
print("RPI Scores for each model:")
print(df[['Model Name'] + [col for col in df.columns if col.startswith('RPI_')]].to_string(index=False))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Data for primary and proposed
models = ['KNN', 'DistanceKNN', 'GeneralizedKNN', 'EuclideanKNN', 'ManhattanKNN',
          'ChebyshevKNN', 'MahalanobisKNN', 'SeuclideanKNN', 'WminkowskiKNN']

primary = {
    'Model Name': models,
    'Accuracy': [0.8155,0.8077,0.8011,0.8155,0.8165,0.8102,0.7747,0.8046,0.8011],
    'Sensitivity': [0.1104,0.1081,0.1374,0.1104,0.1059,0.1081,0.1554,0.1284,0.1374],
    'G-Mean': [0.3231,0.3183,0.3563,0.3231,0.3168,0.3188,0.3717,0.3455,0.3563],
    'F1 Score': [0.1573,0.1493,0.1773,0.1573,0.1526,0.1509,0.1772,0.1701,0.1773]
}
proposed = {
    'Model Name': models,
    'Accuracy': [0.9875,0.9890,0.9891,0.9875,0.9836,0.9824,0.9950,0.9952,0.9891],
    'Sensitivity': [0.9875,0.9890,0.9891,0.9875,0.9836,0.9824,0.9950,0.9952,0.9891],
    'G-Mean': [0.9935,0.9943,0.9943,0.9935,0.9915,0.9908,0.9974,0.9975,0.9943],
    'F1 Score': [0.9870,0.9885,0.9887,0.9870,0.9823,0.9818,0.9949,0.9951,0.9887]
}

metrics = ['Accuracy', 'Sensitivity', 'G-Mean', 'F1 Score']
df_primary = pd.DataFrame(primary).set_index('Model Name')
df_proposed = pd.DataFrame(proposed).set_index('Model Name')

# --- 3. Heatmap of Differences ---
diff = df_proposed[metrics] - df_primary[metrics]

plt.figure(figsize=(10, 6))
sns.heatmap(diff, annot=True, cmap='RdYlGn', center=0, fmt=".3f", linewidths=0.5)
plt.ylabel('Model Name')
plt.xlabel('Metrics')
plt.tight_layout()
plt.show()
```

```python

metric = 'Accuracy'
plt.figure(figsize=(8,5))
plt.plot(models, df_primary[metric], marker='o', label='Baseline ', color='skyblue')
plt.plot(models, df_proposed[metric], marker='s', label='Proposed Method', color='orange')
plt.ylabel(metric, labelpad=15)
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.14), ncol=2, frameon=False)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
